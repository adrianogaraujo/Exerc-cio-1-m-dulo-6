#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
file_path = './sinasc_RO_2019.csv'
df = pd.read_csv(file_path)
print(df.head())
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt

def plot_bar(column_name):
    df[column_name].value_counts().sort_index().plot(kind='bar', figsize=(10, 5))
    plt.title(f'Frequência de {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequência')
    plt.show()

# Variáveis
variables = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

for var in variables:
    plot_bar(var)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
df.groupby('CONSULTAS')['APGAR5'].mean().plot(kind='bar', figsize=(10, 5))
plt.title('APGAR5 médio por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('APGAR5 Médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Variáveis para plotar no painel
variables_panel = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

for ax, var in zip(axs.flatten(), variables_panel):
    df[var].value_counts().sort_index().plot(kind='bar', ax=ax)
    ax.set_title(f'Frequência de {var}')
    ax.set_xlabel(var)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```
